# **Frame the Problem and Look at the Big Picture**
    1. Define the objective in business terms.
To build a model adapted to the prediction of winner's matches.

    2. How will your solution be used?

Predict the outcome of a new tennis match.

    3. What are the current solutions/workarounds (if any)?

There are a few sites where we found already built-solution:

[Tennis Match Prediction with R](https://tennispredict.weebly.com/full-report.html)

[Machine Learning for Professional Tennis Match Prediction and Betting](http://cs229.stanford.edu/proj2017/final-reports/5242116.pdf)

    4. How should you frame this problem (supervised/unsupervised, online/offline,etc.)?

We are in the case of supervised problem because we know the outcomes of matches in the dataset. And in an offline case because we do our predictions locally. As we predict if the given player will be the winner or the loser of the match, we would also classify this problem as a classification task.

    5. How should performance be measured?

We will know if the model is performant if it has a good winrate or not.

    6. Is the performance measure aligned with the business objective?

Yes

    7. What would be the minimum performance needed to reach the business objective?

We need to have more 50% prediction. We fix our minimum performance to 60% prediction. 

It would mean that our model have taken parameters in count that impact the win of the player. And that is exactly what we want !

    8. What are comparable problems? Can you reuse experience or tools?

All problems when there are a winner and a looser given some parameters. Wether in a sport event or in a presidential election. 
It is said in this article [(A machine learning framework for sport result prediction)](https://www.sciencedirect.com/science/article/pii/S2210832717301485), studying the possibilities of sport prediction using machine learning, that Artificial Neural Network (ANN) is the most common model used to make predictions. We will try to reproduce this kind of model.

It is also said in this article [(Horse racing prediction using artificial neural networks)](https://scholar.google.com/scholar_lookup?title=Horse%20racing%20prediction%20using%20artificial%20neural%20networks&publication_year=2010&author=E.%20Davoodi&author=A.%20Khanteymoori) where they compare five type of supervised neural network algorithms. They come out with almost the same results whatever the algorithm they pick. 

    9. Is human expertise available?

A human can also make good predictions (pro-betters).

    10. How would you solve the problem manually?

I would analyse the results of the player #1 in the X last months vs the player #2 and find out some charateristics (player #1 is better in this kind of surface).

This is not how we will try to solve the problem. This method requires to create a "file" for each player and analyse his personal results.

What we will try to is find correlation between the win and attributes (rank, age, hand of the player...)

This is also possible manually but using machine learning will be more effective.

    11. List the assumptions you (or others) have made so far.

We will not have a perfect model because we wan't take the randomness of matches in count

We will use a supervised machine learning algortithm
